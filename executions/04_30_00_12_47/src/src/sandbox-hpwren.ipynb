```python
import numpy as np
from PIL import Image
import requests
from requests.exceptions import HTTPError
from io import BytesIO

def urlToImage(url):
    image = None    
    try:
        response = requests.get(url)
        image = Image.open(BytesIO(response.content))
        response.raise_for_status()
    except HTTPError as http_err:
        print(f'HTTP error occurred: {http_err}')  
    except Exception as err:
        print(f'Other error occurred: {err}')
    return image
```

```python
from matplotlib import pyplot as plt
hpwrenCamsF = hpwrenCams["features"]
hpwrenCamsAtSite = hpwrenCamsF[0]["properties"]["latest-images"]
numImg = len(hpwrenCamsAtSite)
description = hpwrenCamsF[0]["properties"]['description']['name']

for i in range(numImg):
    imageURL = hpwrenCamsAtSite[i][0]["image"]
    
    direction = hpwrenCamsAtSite[i][0]["direction"]
    color = hpwrenCamsAtSite[i][0]["color"]
    diff = hpwrenCamsAtSite[i][0]["diff"]
    colorType = 'Color' if color else 'Monochrome'
    diffType = 'Difference' if diff else 'Original'
        
        
    print("image url: " + imageURL)
    print("description: " + description + ' ' + direction + ' ' + colorType + ' ' + diffType)
    
    image = urlToImage(imageURL)
    plt.imshow(image)
    plt.show()
    print('')
```
